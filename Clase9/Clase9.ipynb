```
import requests
url = 'https://www.eltiempo.com'

response = requests.get(url)
print(response)
```

```
print(response.headers)
print(response.headers['date'])
print(response.text)
```

```
import bs4

completo = bs4.BeautifulSoup(response.text,'html.parser')
noticias = completo.select('.title-container a')

print(noticias)
```

```
titulos = []
links = []
for noticia in noticias:
  titulos.append(noticia.text)
  links.append('https://www.eltiempo.com' + noticia['href'])

print(titulos)
print(links)
dictNoticias = {}
dictNoticias['Noticias'] = titulos
dictNoticias['Enlace'] = links
```

```
import pandas as pd
noticiasElTiempo = pd.DataFrame(dictNoticias)
noticiasElTiempo

noticiasElTiempo.to_csv('Noticias.cvs', index=False)
```
